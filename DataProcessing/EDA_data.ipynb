```python
#import thu vien
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/home/binperdok/KLTN2026/Data/du_lieu_ai_day_du.csv")
```

```python
df[['title', 'content']] = df[['title', 'content']].replace('', np.nan)
df = df.dropna(subset=['title', 'content'])
```

```python
df.info()
```

```python
df = df.assign(label="DEEP DIVE")
```

```python
df.head()
```

```python
df.info()
```

```python
# Xóa hàng có title là null/NaN
df = df.dropna(subset=['content'])

# Xóa hàng có title là empty hoặc "Không thể trích xuất nội dung"
INVALID_TITLES = {"", "Không thể trích xuất nội dung"}
df = df[~df['content'].str.strip().isin(INVALID_TITLES)]

# Xóa hàng có title quá ngắn (dưới 50 ký tự)
df = df[df['content'].str.split().str.len() >= 50]

df = df.reset_index(drop=True)
print(f"Còn lại: {len(df)} hàng")
```
